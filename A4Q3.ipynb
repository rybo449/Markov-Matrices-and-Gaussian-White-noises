```python
#Question 3 Part a)
import numpy as np
import matplotlib.pyplot as plt
A = np.random.normal(loc = 0, scale = 1, size = 1000)
L = [x for x in xrange(1,1001)]
plt.plot(L,A,"r")
plt.ylabel("Random Variable Variance 1")
plt.xlabel("Time")
plt.title("White Noise Gaussian Plot Variance 1")
plt.show()
```

```python
#Question 3 Part b)
X1 = np.random.normal(loc = 0, scale = 1, size = 1000)
X2 = np.random.normal(loc = 0, scale = 1, size = 1000)

L = [x for x in xrange(1,1001)]
A = np.array([ X1, np.ones(1000)])
plt.plot(X1, X2, "r^")
plt.ylabel("X1")
plt.xlabel("X2")
plt.title("White Noise Gaussian Plots of X1 and X2")
plt.grid(True)
w = np.linalg.lstsq(A.T,X2)[0]
line = w[0]*X1 + w[1]
print "Slope:",w[0]
print "Intercept:",w[1]
plt.plot(X1,line)
plt.show()
```
